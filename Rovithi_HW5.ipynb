## Problem 2

Go through KerasCV’s Stable Diffusion tutorial (https://keras.io/guides/keras_cv/generate_images_with_stable_diffusion/), and

b) generate a beautiful drawing of your artistic creation.

HINT: Run the code on Google Collab.

```python
import keras_cv
from keras_cv.models import StableDiffusion
import matplotlib.pyplot as plt
import tensorflow as tf

# Step 3: Use mixed precision for better performance on GPU
tf.keras.mixed_precision.set_global_policy("mixed_float16")

# Step 4: Load model
model = StableDiffusion(img_height=512, img_width=512)
```

```python
# Step 5a: Generate image of a salamander reading a book
prompt_1 = "A beautiful drawing of a salamander reading a book"
image_1 = model.text_to_image(prompt_1, batch_size=1)[0]

plt.imshow(image_1)
plt.axis("off")
plt.title("Salamander Reading a Book")
plt.show()
```

```python
prompt_2 = "A beautiful fantasy artwork of a city in the clouds at sunset"
image_2 = model.text_to_image(prompt_2, batch_size=1)[0]

plt.imshow(image_2)
plt.axis("off")
plt.title("Fantasy City")
plt.show()
```

## Problem 3

What is an attention mechanism? How does it help?

An attention mechanism is a module in a neural network that, for each output element, computes a set of **weights** over all input elements and uses these weights to produce a **weighted sum** of input representations. In practice, it lets the model dynamically “focus” on the most relevant parts of the input rather than treating everything equally.

---

#### How does it help?  
- **Selective focus:** Highlights and uses only the important inputs for each prediction.  
- **Long‑range dependencies:** Directly connects distant elements (e.g., words far apart in a sentence).  
- **Better performance:** Improves accuracy in translation, summarization, question answering, image captioning, etc.  
- **Parallelizable:** Enables faster training (as in Transformer models) compared to sequential RNNs.

## Problem 4: Ethical considerations of deep learning and AI

Read Chapter 21, from “Understanding Deep Learning”, by Simon J. D. Prince, MIT Press (2023), which is accessible online at: from http://udlbook.com.

Which ones are you consider to be the most important? Justify your arguments.

From my perspective, the most important ethical issues in deep learning and AI are **privacy**, **bias and fairness**, and **accountability** / **transparency**.
 
- **Privacy:**  
  AI often relies on huge amounts of data, which sometimes includes personal and sensitive information. Protecting people’s privacy isn’t just a legal obligation—it’s about respecting individuals’ rights and dignity. Without careful handling, there’s a real risk of misuse or unintended exposure of private data, which can have serious consequences.

- **Bias and Fairness:**  
  AI models learn from the data we give them, and if that data reflects existing social biases—whether related to race, gender, or other factors—those biases can end up baked into the model’s decisions. This can lead to unfair or even harmful outcomes, especially in sensitive areas like hiring or healthcare. It’s crucial we recognize this and actively work to minimize bias, so AI can help everyone fairly, not just some groups.

- **Accountability and Transparency:**  
  Many AI models operate like “black boxes”—their inner workings are hard to understand, even for experts. When these models make decisions that impact people’s lives, it’s important to know *how* and *why* those decisions were made. This transparency helps build trust and makes sure someone is responsible if things go wrong.
